```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('./selected.jpg')
plt.imshow(im)
```

```python
# get width and height of the image
h, w = im.shape[:2]
print(h, w)

# crop the image
crop = im[0:h, w//2 - 50:w//2 + 50]
plt.imshow(crop)
```

```python
# detect circles on the cropped image
gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
# gray = cv2.medianBlur(gray, 5)
plt.imshow(gray, cmap='gray')
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=20, minRadius=5, maxRadius=15)
circles = np.uint16(np.around(circles))

# draw circles on the cropped image
for i in circles[0, :]:
    cv2.circle(crop, (i[0], i[1]), i[2], (0, 255, 0), 2)
    cv2.circle(crop, (i[0], i[1]), 2, (0, 0, 255), 3)

plt.figure()
plt.imshow(crop)
```
