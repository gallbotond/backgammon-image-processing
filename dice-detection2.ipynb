```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
im = cv2.imread('./selected.jpg')
plt.imshow(im)
```

```python
# Detect blobs using Difference of Gaussian (DoG) method
blobs = blob_dog(im, max_sigma=30, threshold=.1)
print(blobs)
```

```python
# Convert coordinates to integer
blobs[:, 2] = blobs[:, 2].astype(int)

# Load original image
image_color = cv2.imread('./selected.jpg')
```

```python
print('Shape of blobs:', blobs.shape)
print('First few blobs:', blobs[:5])

# Iterate over each blob and annotate the image
for blob in blobs:
    y, x, size, _ = blob  # Unpack four values
    size = int(size)
    x = int(x)
    y = int(y)
    # Draw a circle around the blob
    cv2.circle(image_color, (x, y), size, (0, 255, 0), 2)
    # Annotate the blob size
    cv2.putText(image_color, str(size), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
```

```python

# Display the image
cv2.imshow('Blobs in image', image_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
